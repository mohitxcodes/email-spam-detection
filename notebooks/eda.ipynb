```python
import pandas as pd
import re

# Load your dataset
df = pd.read_csv("/Users/mohitsingh/code/email-spam/data/email.csv")

# Convert spam/ham to 1/0
df['label'] = df['Category'].map({'spam': 1, 'ham': 0})

# Text cleaning function
def clean_text(text):
    if not isinstance(text, str):
        return ""
    text = text.lower()
    text = re.sub(r"\r|\n", " ", text)
    text = re.sub(r"<.*?>", " ", text)  # remove HTML tags
    text = re.sub(r"http\S+|www\.\S+", " URL ", text)  # mask URLs
    text = re.sub(r"\S+@\S+", " EMAIL ", text)  # mask email addresses
    text = re.sub(r"\s+", " ", text)  # replace multiple spaces
    return text.strip()

# Apply cleaning to your message column
df['cleaned_message'] = df['Message'].apply(clean_text)
df.to_csv("../data/cleaned_emails.csv", index=False)
print(df.head())
```
